```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on="title")
```

```python
movies.head()
```

```python
movies.columns
```

```python
selected_features = ['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew', 'tagline', 'original_language', 'production_companies']
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew', 'tagline', 'original_language', 'production_companies']]
```

```python
movies.head()
```

```python
movies.isna().sum()
```

```python
# replacing the null valuess with null string

for feature in selected_features:
    movies[feature] = movies[feature].fillna('')
```

```python
movies.isna().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies.cast[0]
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:5])
```

```python
movies.cast[0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['production_companies'] = movies['production_companies'].apply(convert)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['production_companies'] = movies['production_companies'].apply(collapse)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tagline'] = movies['tagline'].apply(lambda x:x.split())
movies['title'] = movies['title'].apply(lambda x:x.split())
movies['original_language'] = movies['original_language'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['title'] + movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['tagline'] + movies['original_language'] + movies['production_companies'] 
```

```python
movies.head(1)
```

```python
new = movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew', 'tagline', 'production_companies', 'original_language'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new['title'] = new['title'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)     
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
# converting the text data to feature vectors

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(lowercase=True, stop_words='english')
feature_vectors = vectorizer.fit_transform(new['tags']).toarray()
```

```python
feature_vectors.shape
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
vectorizer.get_feature_names()
```

```python
new[new['title'] == 'Spectre'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
